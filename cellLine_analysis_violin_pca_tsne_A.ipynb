```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mannwhitneyu

# Load the dataset
file_path = '5 (2) (6).xlsx'
data = pd.read_excel(file_path)

# Step 1: Extract 'Cell Line' and 'Day' from the 'Cell_Lines' column
data['Cell Line'] = data['Cell_Lines'].apply(lambda x: x.split(' ')[0])  # Extract cell line
data['Day'] = data['Cell_Lines'].apply(lambda x: x.split(' ')[-1])       # Extract day

# Define features and grouping variable
features = ['Colony Elongation', 'Colony Diameter', 'Lumen Volume']
group = "Cell Line"
unique_groups = data[group].unique()

# Function to create a violin plot and print p-values for one-sided test against 'MCF10A'
def plot_violin(feature, figsize=(10, 6), color="muted"):
    plt.figure(figsize=figsize)

    # Update plot parameters globally
    plt.rcParams.update({
        'font.family': 'serif',
        'font.serif': ['Times New Roman'],
        'font.size': 20,
        'font.weight': 'bold',
        'axes.titlesize': 20,
        'axes.labelsize': 20,
        'xtick.labelsize': 20,
        'ytick.labelsize': 20,
        'axes.titleweight': 'bold',
        'axes.labelweight': 'bold',
        'legend.fontsize': 17,
        'legend.title_fontsize': 17
    })
    
    # Create the violin plot
    sns.violinplot(x=group, y=feature, data=data, palette=color, inner="point")
    plt.ylabel(feature)
    # plt.xlabel(group)

    # Filter pairs to compare each cell line with 'MCF10A'
    pairs = [(unique_groups[i], unique_groups[j]) for i in range(len(unique_groups)) for j in range(i + 1, len(unique_groups))]
    y_max = data[feature].max()
    y_min = data[feature].min()
    offset = 1
    line_spacing = offset * 2

    print(f"P-values for {feature}:")
    for pair_idx, (g1, g2) in enumerate(pairs):
        group1_data = data[data[group] == g1][feature].dropna()
        group2_data = data[data[group] == g2][feature].dropna()
        stat, p_value = mannwhitneyu(group1_data, group2_data, alternative='two-sided')
        
        # Print p-values
        print(f"{g2} vs {g1}: p-value = {p_value:.5f}")

        # Add significance stars based on p-value
        if p_value < 0.001:
            stars = "***"
        elif p_value < 0.01:
            stars = "**"
        elif p_value < 0.05:
            stars = "*"
        else:
            continue

        x1 = unique_groups.tolist().index(g1)
        x2 = unique_groups.tolist().index(g2)
        y = y_max + offset + (pair_idx * line_spacing)
        plt.plot([x1, x2], [y, y], color='black', linewidth=1.5)
        plt.text((x1 + x2) * 0.5, y + 0.3, stars, ha='center', va='bottom', color='red', fontsize=12)

    plt.xticks(rotation=0)
    plt.ylim(-1, y_max + (len(pairs) + 1) * line_spacing)
    plt.tight_layout()
    plt.show()

# Plot for Colony Elongation
plot_violin('Colony Elongation', figsize=(9, 6), color="pastel")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mannwhitneyu

# Load the dataset
file_path = '5 (2) (2).xlsx'
data = pd.read_excel(file_path)

# Step 1: Extract 'Cell Line' and 'Day' from the 'Cell_Lines' column
data['Cell Line'] = data['Cell_Lines'].apply(lambda x: x.split(' ')[0])  # Extract cell line
data['Day'] = data['Cell_Lines'].apply(lambda x: x.split(' ')[-1])       # Extract day

# Define features and grouping variable
features = ['Colony Elongation', 'Colony Diameter', 'Lumen Volume']
group = "Cell Line"
unique_groups = data[group].unique()

# Normalize each feature per cell line
data[features] = data.groupby('Cell Line')[features].transform(lambda x: (x - x.mean()) / x.std())

# Function to create a violin plot and print p-values for one-sided test against 'MCF10A'
def plot_violin(feature, figsize=(10, 6), color="muted"):
    plt.figure(figsize=figsize)

    # Update plot parameters globally
    plt.rcParams.update({
        'font.family': 'serif',
        'font.serif': ['Times New Roman'],
        'font.size': 20,
        'font.weight': 'bold',
        'axes.titlesize': 20,
        'axes.labelsize': 20,
        'xtick.labelsize': 20,
        'ytick.labelsize': 20,
        'axes.titleweight': 'bold',
        'axes.labelweight': 'bold',
        'legend.fontsize': 17,
        'legend.title_fontsize': 17
    })
    
    # Create the violin plot
    sns.violinplot(x=group, y=feature, data=data, palette=color, inner="point")
    plt.ylabel(f"{feature} (Normalized per Cell Line)")
    plt.xlabel(group)

    # Filter pairs to compare each cell line with 'MCF10A'
    reference_group = 'MCF10A'
    pairs = [(unique_groups[i], unique_groups[j]) for i in range(len(unique_groups)) for j in range(i + 1, len(unique_groups))]
    y_max = data[feature].max()
    y_min = data[feature].min()
    offset = 0.5
    line_spacing = offset * 2

    print(f"P-values for {feature} (One-sided test against {reference_group}):")
    for pair_idx, (g1, g2) in enumerate(pairs):
        group1_data = data[data[group] == g1][feature].dropna()
        group2_data = data[data[group] == g2][feature].dropna()
        stat, p_value = mannwhitneyu(group1_data, group2_data, alternative='two-sided')
        
        # Print p-values
        print(f"{g2} vs {g1}: p-value = {p_value:.5f}")

        # Add significance stars based on p-value
        if p_value < 0.001:
            stars = "***"
        elif p_value < 0.01:
            stars = "**"
        elif p_value < 0.05:
            stars = "*"
        else:
            continue

        x1 = unique_groups.tolist().index(g1)
        x2 = unique_groups.tolist().index(g2)
        y = y_max + offset + (pair_idx * line_spacing)
        plt.plot([x1, x2], [y, y], color='black', linewidth=1.5)
        plt.text((x1 + x2) * 0.5, y + 0.3, stars, ha='center', va='bottom', color='red', fontsize=12)

    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()

# Plot for Lumen Volume
plot_violin('Lumen Volume', figsize=(10, 5), color="muted")
```
