```python
import os
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from skimage import io, exposure
from skimage.feature import blob_log

def process_image(file_path, min_sigma, max_sigma):
    """
    Processes a single image for blob detection using Laplacian of Gaussian (LoG).
    
    Parameters:
    - file_path (str): The path to the image file.
    - min_sigma (float): Minimum standard deviation for Gaussian kernel.
    - max_sigma (float): Maximum standard deviation for Gaussian kernel.
    
    Returns:
    - Data dictionary containing file name, detected blob count, and blob details.
    """
    if not os.path.exists(file_path):
        print(f"Error: The file '{file_path}' does not exist.")
        return

    # Load the image
    image = io.imread(file_path)
    
    # Preprocess the image (contrast stretching)
    low_value, high_value = np.percentile(image, (50, 98))
    stretched_image = exposure.rescale_intensity(image, in_range=(low_value, high_value))
    
    # Perform blob detection using LoG
    blobs_log = blob_log(stretched_image, min_sigma=min_sigma, max_sigma=max_sigma, 
                         threshold=0.005, overlap=0.25, log_scale=True)
    blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)  # Adjust radii

    # Visualization
    fig, ax = plt.subplots(1, 1, figsize=(6, 6))
    ax.imshow(stretched_image, cmap='gray')
    for blob in blobs_log:
        y, x, r = blob
        c = plt.Circle((x, y), r, color='yellow', linewidth=2, fill=False)
        ax.add_patch(c)
    
    ax.set_axis_off()
    plt.show()

    # Prepare data for output
    data = {
        'file_name': os.path.basename(file_path),
        'cell_count': len(blobs_log),
        'blob_details': blobs_log
    }
    return data
```

```python
process_image("E7_DAPI_1.tif", min_sigma=4, max_sigma=7)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from skimage import io, exposure
from skimage.feature import blob_log
import base64
from io import BytesIO

def process_image_cell_count(file_path, min_sigma, max_sigma):
    """
    Processes a single image for blob detection using Laplacian of Gaussian (LoG)
    and returns the image as a Base64 string.
    
    Parameters:
    - file_path (str): The path to the image file.
    - min_sigma (float): Minimum standard deviation for Gaussian kernel.
    - max_sigma (float): Maximum standard deviation for Gaussian kernel.
    
    Returns:
    - Data dictionary containing file name, detected blob count, blob details, 
      and the processed image in Base64 format.
    """
    if not os.path.exists(file_path):
        print(f"Error: The file '{file_path}' does not exist.")
        return

    # Load the image
    image = io.imread(file_path)
    
    # Preprocess the image (contrast stretching)
    low_value, high_value = np.percentile(image, (50, 98))
    stretched_image = exposure.rescale_intensity(image, in_range=(low_value, high_value))
    
    # Perform blob detection using LoG
    blobs_log = blob_log(stretched_image, min_sigma=min_sigma, max_sigma=max_sigma, 
                         threshold=0.005, overlap=0.25, log_scale=True)
    blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)  # Adjust radii

    # Visualization and converting to Base64
    fig, ax = plt.subplots(1, 1, figsize=(6, 6))
    ax.imshow(stretched_image, cmap='gray')
    for blob in blobs_log:
        y, x, r = blob
        c = plt.Circle((x, y), r, color='yellow', linewidth=2, fill=False)
        ax.add_patch(c)
    ax.set_axis_off()

    # Convert the plot to a Base64 string
    buffer = BytesIO()
    plt.savefig(buffer, format="png", bbox_inches='tight', pad_inches=0)
    buffer.seek(0)
    img_base64 = base64.b64encode(buffer.getvalue()).decode('utf-8')
    plt.close(fig)

    # Prepare data for output
    data = {
        'file_name': os.path.basename(file_path),
        'cell_count': len(blobs_log),
        'image_base64': img_base64
    }
    return data
```

```python
process_image_cell_count("E7_DAPI_1.tif", min_sigma=4, max_sigma=7)
```
